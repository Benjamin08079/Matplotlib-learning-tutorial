# Matplotlib Tutorial

-it is an open source drawing library which supports rich drawing types.

-it is used to draw 2D and 3D graphics.

- you can understand your data easily by visualizing it with the help of matplotlib!

- you can generate plots,histogram, bar charts and many other charts with just a few lines of code

- #types of charts
-Bar chart, scatter, pie chart, histogramm, line charts, Area graph.

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8, 0.8]) # control the left, right, width, heigth of the canvas ( from 0 to 1)

axis.plot(x,y, 'r')
```

```python
# again we can draw subgraphs
fig, axes = plt.subplots(nrows = 1, ncols=2) # Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# we can also draw a picture, or graph, inside another graph

# control the left, right, width, height of the canvas ( from 0 to 1)

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

# Other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2, lw=2)

axes[2].set_title("bar")
axes[2].bar(n, n**2, align="center", alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)                  
                  
```

```python
"""Draw a radar chart"""

fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, 1, 1], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t,t, color='blue', lw=3)
```

```python
"""Draw a histogram"""

n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("cumulative detailed histogram")
axes[1].hist(n, cumulative=True, bins=8)
```

```python
# Draw a contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt


delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(x-1)**2-(Y-1)**2)
Z = (Z1 -Z2) * 2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour map')
```

```python
# Add labels

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
# Explode
#pull the "Apples" wedge 0.2 from the center of the pie

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
#Add a shadow

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```
